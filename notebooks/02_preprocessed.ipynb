```python
import sys
import pandas as pd

sys.path.append("..")
from src.eda import cap_processing_fee, ResidenceTypeImputer, cap_gst, cap_net_disbursement, CategoryCleaner
from src.feature_engineering import create_loan_to_income, create_delinquency_ratio, create_avg_dpd_per_delinquency
```

```python
train_df = pd.read_parquet("../DATA/raw/train_df_raw.parquet")
test_df = pd.read_parquet("../DATA/raw/test_df_raw.parquet")

train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
train_df.columns[train_df.isna().sum()>0]
```

```python
res_imput = ResidenceTypeImputer()
res_imput.fit(train_df)
train_df = res_imput.transform(train_df)

test_df = res_imput.transform(test_df)
```

```python
train_df.columns[train_df.isna().sum()>0]
```

```python
train_df = cap_processing_fee(train_df)
test_df = cap_processing_fee(test_df)
train_df = cap_gst(train_df)
test_df = cap_gst(test_df)
train_df = cap_net_disbursement(train_df)
test_df = cap_net_disbursement(test_df)
```

```python
cat_clean = CategoryCleaner()
train_df = cat_clean.transform(train_df)
test_df = cat_clean.transform(test_df)
```

```python
train_df.shape, test_df.shape
```

```python
# Feature Engineering

train_df = create_loan_to_income(train_df)
test_df = create_loan_to_income(test_df)
train_df = create_delinquency_ratio(train_df)
test_df = create_delinquency_ratio(test_df)
train_df = create_avg_dpd_per_delinquency(train_df)
test_df = create_avg_dpd_per_delinquency(test_df)


```

```python
train_df.shape, test_df.shape
```

```python
train_df.to_parquet("../DATA/proprocessed_data/train_df_preprocessed.parquet")
test_df.to_parquet("../DATA/proprocessed_data/test_df_preprocessed.parquet")
```
